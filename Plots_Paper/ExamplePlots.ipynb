```julia
using Plots
gr()  # backend

x = 0:0.1:10
y1 = sin.(x)
y2 = 100 .* cos.(x)

p = plot(x, y1,
         label="sin(x)",
         xlabel="X",
         ylabel="Left Y",
         color=:blue)

plot!(twinx(), x, y2,
      label="100*cos(x)",
      ylabel="Right Y",
      color=:red)

display(p)
```

```julia
using Plots
gr()

x1 = 0:0.1:10
x2 = 0:0.1:10
y  = sin.(x1)

p = plot(x1, y,
         label="sin(x)",
         xlabel="Bottom X",
         ylabel="Y",
         color=:blue)

plot!(twiny(), x2, y .* 2,
      label="2*sin(x)",
      xlabel="Top X",
      color=:red)

display(p)
```

```julia
using Plots
gr()

A = rand(50, 50)

p = heatmap(A,
            xlabel="X",
            ylabel="Left Y",
            colorbar=true)

# Overlay a line with a different Y-scale
y_overlay = range(1, 50, length=50)
x_overlay = 1:50

plot!(twinx(), x_overlay, 10 .* rand(50),
      color=:white,
      label="Overlay",
      ylabel="Right Y")

display(p)
```

```julia
using Plots
gr()

x = 1:0.1:10
y1 = x.^2
y2 = 10 .^ x   # strictly positive (required for log scale)

p = plot(x, y1,
         xlabel="X",
         ylabel="Left Y (linear)",
         label="x^2",
         color=:blue)

plot!(twinx(), x, y2,
      ylabel="Right Y (log10)",
      yscale=:log10,
      label="10^x",
      color=:red)

display(p)
```

```julia
using Plots
gr()

x = 1:0.1:10
y1 = 10 .^ (0.5 .* x)
y2 = 10 .^ x

p = plot(x, y1,
         xlabel="X",
         ylabel="Left Y (log)",
         yscale=:log10,
         label="10^(0.5x)",
         color=:blue)

plot!(twinx(), x, y2,
      ylabel="Right Y (log)",
      yscale=:log10,
      label="10^x",
      color=:red)

display(p)
```

```julia
using Plots
gr()

A = rand(50, 50) .+ 1e-3  # avoid zeros if using log

p = heatmap(A,
            xlabel="X",
            ylabel="Left Y",
            yscale=:log10,     # log scale on heatmap axis
            colorbar=true)

plot!(twinx(), 1:50, 10 .^ (0.02 .* (1:50)),
      yscale=:log10,
      ylabel="Right Y (log)",
      color=:white,
      label="overlay")

display(p)
```

```julia
using Plots
gr()

nx, ny = 100, 80
A = rand(ny, nx)

x = 1:nx
y = 1:ny

# Main heatmap
p = heatmap(x, y, A,
    xlabel = "X",
    ylabel = "Left Y",
    colorbar = true,
    framestyle = :box,
    aspect_ratio = :equal,
    right_margin = 15Plots.mm  # create space BEFORE colorbar
)

# Lock limits
xlims!(minimum(x), maximum(x))
ylims!(minimum(y), maximum(y))

# Add right Y axis
plot!(twinx(),
    fill(first(x), length(y)), y,   # dummy invisible vertical line
    ylabel = "Right Y",
    ylims = (minimum(y), maximum(y)),
    xlims = (minimum(x), maximum(x)),
    legend = false,
    grid = false,
    linealpha = 0
)

display(p)
```
